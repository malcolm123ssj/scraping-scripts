```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
import pyautogui as auto
import time

auto.PAUSE =2.0
```

```python
driver = webdriver.Chrome(executable_path="/usr/bin/chromedriver")
driver.get("https://www.deccanherald.com/tag/mumbai")
auto.click(579 ,24 ,clicks=2)

time_end =time.time() +60*1
time.sleep(5)
auto.moveTo(1694,808,duration=0.5)

while time.time() < time_end:
    continue
     

res = driver.execute_script("return document.documentElement.outerHTML")
soup =bs(res,'lxml')

a =soup.find_all('a',href=True)

url =[]
for i in a:
    if '/city/' in i.get('href') or '/national/' in i.get('href') or '/photos/' in i.get('href'):
        url.append(i.get('href'))
```

```python
len(url)
```

```python
base = 'https://www.deccanherald.com'

news_crime =[]

for i in url:
    req = requests.get(base+i)
    soup = bs(req.text,"lxml")
    
    title = soup.find_all('h2' ,class_='f-left sanspro-b')
    title = title[0].text
    
    tags_even = soup.find_all('div' ,class_='field-item even')
    tags_odd = soup.find_all('div' ,class_='field-item odd')
    
    odd = ""
    for i in tags_odd:
        odd+=i.text +','
    
    even = ""
    for i in range (len(tags_even)-1 ,1 ,-1):
        even+=tags_even[i].text +','
    
    tags =even+odd
        
    text=""
    for paragraph in soup.find_all('p'):
        text+=paragraph.text
    
    ul = soup.find('ul' ,class_='crud-items')
    li = ul.findChildren("li" , recursive=False)
    date = li[0].text
    date = date.split(',')
    
    temp =[]
    
    temp.append(title)
    temp.append(text)
    temp.append(tags)
    temp.append(date[0])
    
    news_crime.append(temp)
```
