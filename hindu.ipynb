```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# scraping hindu newspaper archives

req = requests.get('https://www.thehindu.com/archive/',timeout=240)
soup = bs(req.text, 'lxml')
```

```python
months = soup.find_all('ul' ,class_='archiveMonthList')
```

```python
li = []
for i in months:
    li.append(i.findChildren("li" , recursive=False))
```

```python
li
```

```python
links =[]
for i in li:
    for j in i:
        links.append(j.findChildren("a" , recursive=False))
```

```python
url =[]
for i in links:
     url.append(i[0].get('href'))
```

```python
link =[]
for i in url:
    if 'web' not in i:
        link.append(i)
```

```python
type(link[0])
```

```python
req = requests.get('https://www.thehindu.com/archive/print/2000/01/',timeout=240)
soup = bs(req.text, 'lxml')
```

```python
tr = soup.find_all('tr')
```

```python
td = []
for i in tr:
    td.append(i.findChildren("td" , recursive=False))
```

```python
a =[]
for i in td:
    for j in i:
        a.append(j.findChildren("a" , recursive=False))
```

```python
days = []
for i in a:
    if len(i)!=0:
        days.append(i[0].get('href'))
```

```python
news =[]

for url in link_day:
    req = requests.get(url,timeout=240)
    soup = bs(req.text, 'lxml')
    
    links = soup.find_all('ul', class_='archive-list')
    
    temp =[]
    for i in links:
        temp.append(i.findChildren("li" , recursive=False))
    
    x = []
    for i in temp:
        for j in i:
            x.append(j.findChildren("a" , recursive=False))
    
    for i in x:
        news.append(i[0].get('href'))
```

```python
h =[]
for i in link_day:
    num =int(i[-11:-7])
    if num ==2018:
        h.append(i)
```

```python
h
```

```python
g =[]
for i in h:
    num =int(i[-6:-4])
    if num >6 and num<=9:
        g.append(i)
```

```python
g
```

```python
news =[]

for url in g:
    req = requests.get(url,timeout=240)
    soup = bs(req.text, 'lxml')
    
    links = soup.find_all('ul', class_='archive-list')

    temp =[]
    for i in links:
        temp.append(i.findChildren("li" , recursive=False))
    
    x = []
    for i in temp:
        for j in i:
            x.append(j.findChildren("a" , recursive=False))
    
    for i in x:
        news.append(i[0].get('href'))
```
