```python
import requests
from bs4 import BeautifulSoup as bsoup
no = 2
skills =[]
for i in range(2,7):
    base = 'http://www.jobspider.com/job/resume-search-results.asp/words_software+developer/searchtype_1/page_'
    url =base+str(no)
    req = requests.get(url ,timeout=240)
    soup = bsoup(req.text, 'lxml')
    no+=1
    links = soup.find_all('a', href=True)
    base_url = 'http://www.jobspider.com'
    resume =[]
    for i in links:
        if 'view-resume' in i.get('href'):
            resume.append(base_url+i.get('href'))

    for i in resume:
        res = requests.get(i, timeout=240)
        soup = bsoup(res.text, 'lxml')
        data = soup.find_all('font')

        for i in data:
            if 'Skills' in i.text:
                skills.append(i.text)
```

```python
for i in skills:
    print(i)
```

```python
with open('resume.txt', 'w') as f:
    for i in skills:
        f.write("%s\n" % i)
```
